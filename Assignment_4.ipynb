Sin and Cos Curves:

```
import matplotlib.pyplot as plt
import math
 
x = [0.1*i for i in range(64)]
sin = [math.sin(i) for i in x]
cos = [math.cos(i) for i in x]
 
plt.plot(x, sin, label='sin(x)', color='blue', linewidth=2.5)
plt.plot(x, cos, label='cos(x)', color='red', linewidth=0.5)
plt.ylabel('y')
plt.xlabel('x')
plt.legend(loc='upper right')
plt.show()
```

$ X^2 $ Curve <br>
Exponential Curve ( $e ^ x $) <br>
Logarithmic Curve

```
x = [0.1*i for i in range(1, 64)]
x_square = [i*i for i in x]
exponential = [math.exp(i) for i in x]
log = [math.log10(i) for i in x]
 
plt.plot(x, x_square, label='$ x^2 $', color='blue', linewidth=2.5)
plt.plot(x, exponential, label='$ e^x $', color='red', linewidth=2.5)
plt.plot(x, log, label='log', color='green', linewidth=2.5)
plt.ylabel('y')
plt.xlabel('x')
plt.legend(loc='upper left')
plt.show()
```
